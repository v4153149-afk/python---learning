```python
tips = sns .load_dataset("tips")
print(type(tips))
print(tips.head())
```

```python
#RELATIONAL 
import seaborn as sns
sns.set_theme(style="ticks")

sns.relplot(
      data=tips,
      x="total_bill",
      y="tip",
      size="size",
      hue="smoker",
      style="smoker"
)
```

```python
sns.relplot(
      data=tips,
      x="total_bill",
      y="tip",
      size="size",
      hue="smoker",
      style="smoker",
      kind="line"
)
```
